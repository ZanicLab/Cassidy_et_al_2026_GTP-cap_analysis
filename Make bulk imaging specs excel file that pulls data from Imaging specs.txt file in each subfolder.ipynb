```python
import os
import re
import pandas as pd

# Directories to check
directories_to_check = [
    r"C:\Users\annao\OneDrive - Vanderbilt\Desktop\To_be_analyzed\VF MCAK Data from 2019"
]

output_data = []

for folder_path in directories_to_check:
    folder_name = os.path.basename(folder_path)
    specs_path = os.path.join(folder_path, "Imaging specs.txt")
    
    if not os.path.isfile(specs_path):
        print(f"Missing file: {specs_path}")
        continue

    with open(specs_path, 'r') as f:
        content = f.read()

    # Split into blocks using lines of dashes
    blocks = [b.strip() for b in content.split('---------------------------') if b.strip()]
    
    for block in blocks:
        lines = block.splitlines()
        filename = lines[0].strip() if lines else None

        m_frames = re.search(r"Total\s+Frames:\s*(\d+)", block, re.I)
        m_duration = re.search(r"Total\s+Imaging\s+Duration:\s*([\d.]+)", block, re.I)
        m_spf = re.search(r"Frame\s+Rate:\s*([\d.]+)\s*(?:frames?/s|s/frame)", block, re.I)  # incorrectly labeled but correctly formatted
        m_pixel = re.search(r"Pixel\s+Size:\s*([\d.]+)", block, re.I)

        data = {
            "Subfolder": folder_name,
            "Filename": filename,
            "Total Frames": int(m_frames.group(1)) if m_frames else None,
            "Total Imaging Duration (s)": float(m_duration.group(1)) if m_duration else None,
            "Frame Rate (s/frame)": float(m_spf.group(1)) if m_spf else None,
            "Pixel Size (um/pixel)": float(m_pixel.group(1)) if m_pixel else None
        }

        output_data.append(data)

# Save to Excel
output_df = pd.DataFrame(output_data)
output_path = r"C:\Users\annao\OneDrive - Vanderbilt\Desktop\To_be_analyzed\OldMCAK-Pooled_Imaging_Specs_Summary.xlsx"
output_df.to_excel(output_path, index=False)
print(f"Excel file saved: {output_path}")
```
